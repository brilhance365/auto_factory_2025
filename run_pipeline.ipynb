```python
# ==========================================
# RUN_PIPELINE - ORQUESTRADOR DEFINITIVO
# ==========================================
import time

# NOMES EXATOS CONFORME A SUA IMAGEM:
# 1. 02_bronze_orders (Supondo que este faz a ingestão)
# 2. 03_silver_orders
# 3. 04_gold_orders
NOTEBOOKS = [
    ("Bronze", "./02_bronze_orders"),
    ("Silver", "./03_silver_orders"),
    ("Gold",   "./04_gold_orders"),
]

def run_layer(label, path, timeout_seconds=1200):
    print(f"Iniciando {label}: {path}...", end="")
    try:
        start = time.time()
        # O prefixo "./" garante que ele procure na mesma pasta do run_pipeline
        dbutils.notebook.run(path, timeout_seconds)
        duration = round(time.time() - start, 2)
        print(f" [OK] em {duration}s")
        return True
    except Exception as e:
        print(f" [FALHOU]")
        print(f"ERRO: {str(e)}")
        return False

# --- EXECUÇÃO DO FLUXO MEDALLION ---
print(f"--- START PIPELINE: {time.strftime('%Y-%m-%d %H:%M:%S')} ---")

success = True
for label, path in NOTEBOOKS:
    if not run_layer(label, path):
        print(f"\n[INTERROMPIDO] Falha crítica na camada {label}.")
        success = False
        break

if success:
    print("\n" + "="*35)
    print("PIPELINE AUTO_FACTORY_2025 CONCLUÍDO!")
    print(f"Finalizado em: {time.strftime('%H:%M:%S')}")
    print("="*35)
```
